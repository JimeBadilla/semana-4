```python
dfAutos[["normalized", "peak"]] = dfAutos[["normalized", "peak"]].astype("int")
dfAutos[["bore", "stroke", "price"]] = dfAutos[["bore", "stroke", "price"]].astype("float")
```

```python
dfAutos["Higway-L/100km"] = 235/dfAutos["Higway"]
dfAutos.head(5)
```

<h3 Style="color:#FF7F50">3.3 Normalizacion de Datos </h3>

```python
dfAutos.info
```

```python
dfTemp = dfAutos[['length', 'width', 'heingt']] 
dfTemp.head(5) 
```

```python
dfAutos['length'] = dfAutos['length']/dfAutos['length'].max() 
dfAutos['width'] = dfAutos['width']/dfAutos['width'].max() 
dfAutos['heingt'] = dfAutos['heingt']/dfAutos['heingt'].max() 
```

```python
dfTemp = dfAutos[['length', 'width', 'heingt']] 
dfTemp.head(5) 
```

<h3 Style="color:#DC143C">3.4 Bining </h3>

```python
dfAutos.head(5)
```

```python
dfAutos[['horsepower']] = dfAutos[['horsepower']].astype("int")
dfAutos.info() 
```

```python
cat=np.linspace(min(dfAutos['horsepower']), max(dfAutos['horsepower']), 4)
cat
```

```python
nombres = ["bajo", "medio", "alto"]
dfAutos["horsepower-bins"] = pd.cut(dfAutos['horsepower'], bins = cat, labels = nombres, include_lowest=True)
dfAutos[["horsepower", "horsepower-bins"]].head(5)
```

```python
dfAutos["horsepower-bins"].value_counts()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.hist(dfAutos["horsepower-bins"])
plt.xlabel("horsepower")
plt.title("Distribucion de Autos segun caballos de fuerza")
plt.show()
```

<h2 Style="color:#FFD700">4. Visualizacion de Datos</h2>

<p> Que se hace en esta etapa?</p>

<ol>
    <li>Responder las preguntas de negocio</li>
    <li>Liberar Conocimiento</li>
    <li>Generar informe al negocio</li>
</ol>
  


```python
gf1 = dfAutos.groupby(["horsepower-bins"]).count()
gf1 =  gf1.iloc[:,:1]
gf1
```

```python
eje_x = ("bajo", "medio", "alto")
eje_y = gf1['symboling'].tolist()

##Creamos la grafica
plt.bar(eje_x, eje_y)

##Leyenda en el eje y
plt.ylabel('Cantidad de Autos')

##Leyenda en el eje x
plt.xlabel('Caballos de Fuerza')

##Titulo del Grafico
plt.title("Autos agrupados por caballos de Fuerza")

##Tamano del grafico
plt.figure(figsize=(15, 30)) 
plt.show()
```
