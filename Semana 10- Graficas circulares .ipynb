<h3 Style="color:#DC143C">3.4 Bining </h3>

```python
dfAutos.head(5)
```

```python
dfAutos[['horsepower']] = dfAutos[['horsepower']].astype("int")
dfAutos.info() 
```

```python
cat=np.linspace(min(dfAutos['horsepower']), max(dfAutos['horsepower']), 4)
cat
```

```python
nombres = ["bajo", "medio", "alto"]
dfAutos["horsepower-bins"] = pd.cut(dfAutos['horsepower'], bins = cat, labels = nombres, include_lowest=True)
dfAutos[["horsepower", "horsepower-bins"]].head(5)
```

```python
dfAutos["horsepower-bins"].value_counts()
```

```python
gf1 = dfAutos.groupby(["horsepower-bins"]).count()
gf1 =  gf1.iloc[:,:1]
gf1
```

```python
eje_x = ("bajo", "medio", "alto")
eje_y = gf1['symboling'].tolist()

##Creamos la grafica
plt.bar(eje_x, eje_y)

##Leyenda en el eje y
plt.ylabel('Cantidad de Autos')

##Leyenda en el eje x
plt.xlabel('Caballos de Fuerza')

##Titulo del Grafico
plt.title("Autos agrupados por caballos de Fuerza")

##Tamano del grafico
plt.figure(figsize=(15, 30)) 
plt.show()
```

```python
gf2 = dfAutos.groupby(['make']).count() 
gf2 =  gf2.iloc[:,:1]
gf2
```

```python
eje_x = ("alfa-romero", "audi", "bmw", "chevrolet", "dodge", "honda", "isuzu", "jaguar", "mazda", "mercedes-benz", "mercury",
"mitsubishi", "nissan", "peugot", "plymouth", "porsche", "saab", "subaru", "toyota", "volkswagen", "volvo")
eje_y = gf2['symboling'].tolist()


plt.rcdefaults()
figure, ax = plt.subplots() 


# Este se alimenta de la lista denominada eje x
y_pos = np.arange(len(eje_x))


ax.barh(y_pos, eje_y, align='center', color= "#FF8C00", alpha=0.5 )

ax.set_yticks(y_pos, labels=eje_x, color = "black") 
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Cantidad de Vehiculos').set_color('#1E90FF')
ax.set_title('Cantidad de Vehiculos segun fabricante') 

plt.show()
```

```python
gf3 = dfAutos.groupby(['doors']).count() 
gf3 =  gf3.iloc[:,:1]
gf3
```

```python
etiqueta = "Four", "Two"
sizes = gf3["symboling"].tolist()

separacion = (0, 0.1)  # para separar el grafico

figl, axl = plt.subplots()

axl.pie(sizes, explode=separacion, labels=etiqueta, autopct='%1.1f%%', shadow=True, startangle=90)
axl.axis('equal') 
plt.show
```

```python
gf4 = dfAutos.groupby(['style']).count() 
gf4 =  gf4.iloc[:,:1]
gf4
```

```python
fig, ax = plt.subplots(figsize=(8, 4), subplot_kw=dict(aspect="equal"))


data = gf4 ['symboling'].tolist()

estilo= ['convertible', 'Hardtop', 'Hatchback', 'Sedan', 'Wagon']


def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return "{:.1f}%".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, estilo,
          title="Estilo",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1.7))

plt.setp(autotexts, size=6, weight="bold")

ax.set_title("Autos segun estilo") 

plt.show()
```

```python
fig, ax = plt.subplots ()

size=0.3
vals= data

cmap = plt.colormaps["tab20c"]
outer_colors = cmap(np.arange(4)*5)

ax.pie(vals, radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.set(aspect="equal", title='Autos segun Estilo')
plt.show()
```
